```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def ends(df, x=5):
    return df.head(x).append(df.tail(x))
setattr(pd.DataFrame,'ends',ends)
setattr(pd.Series,'ends',ends)
```

```python
PATH = './../'
PATH_DATA_RAW = PATH + 'data/raw/'
PATH_DATA_INTERMEDIATE = PATH + 'data/intermediate/'
PATH_DATA_PROCESSED = PATH + 'data/processed/'
```

# Loading data

```python
filepath = 'https://intl-atlas-downloads.s3.amazonaws.com/country_hsproduct4digit_year.csv.zip'

df = pd.read_csv(filepath,
                 dtype={"year": int, "hs_product_code": str})
```

```python
df.ends(3)
```

```python
df.shape
```

# Some cleaning

```python
# drop some years
years2drop = [yr for yr in range(df.year.min(), df.year.max()+1) if yr%5!=0]

# revome the same countries we removed in our paper
# see supplementary information in https://arxiv.org/abs/1812.03534
countries2drop = ['SRB', 'CUW', 'SXM', 'BLM', 'BES', 'ETH', 'GEO', 'MMR', 'SEN', 'TKM', 'TZA', 'UGA', 
                  'AFG', 'ASM', 'AND', 'AIA', 'ATA', 'ATG', 'ARM', 'ABW', 'BHS', 'BHR', 'BRB', 'BLZ', 'BEN', 'BMU', 'BTN', 
                  'BVT', 'IOT', 'VGB', 'BRN', 'BFA', 'BDI', 'CPV', 'CYM', 'CAF', 'TCD', 'CXR', 'CCK', 'COM', 'COK', 'CYP', 
                  'DJI', 'DMA', 'GNQ', 'ERI', 'FLK', 'FRO', 'FJI', 'GUF', 'PYF', 'ATF', 'GMB', 'GIB', 'GRL', 'GRD', 'GUM', 
                  'GNB', 'GUY', 'HTI', 'HMD', 'VAT', 'ISL', 'IRQ', 'KIR', 'LAO', 'LSO', 'LUX', 'MAC', 'MWI', 'MDV', 'MLT', 
                  'MHL', 'MTQ', 'MRT', 'MUS', 'MYT', 'FSM', 'MNG', 'MNE', 'MSR', 'NRU', 'NPL', 'ANT', 'NCL', 'NER', 'NIU', 
                  'NFK', 'MNP', 'PCI', 'PLW', 'PSE', 'PCN', 'RWA', 'SHN', 'KNA', 'LCA', 'SPM', 'VCT', 'WSM', 'SMR', 'STP', 
                  'SYC', 'SLE', 'SLB', 'SOM', 'SGS', 'SSD', 'SUR', 'SWZ', 'TWN', 'TLS', 'TGO', 'TKL', 'TON', 'TCA', 'TUV', 
                  'UMI', 'VUT', 'VIR', 'WLF', 'ESH', 'YEM']

# remove ores, minerals, coal, and petroleum and oil products
products2drop = ['2601', '2603', '2607', '2608', '2612', 
                 '2701', '2709', '2710', '2711', '2713', '2715',  
                 '7102', '7106', '7108', '7110', 
                 '7402', '7403', 
                 '7901']
```

```python
df = df[~((df['year'].isin(years2drop)) | 
           (df['location_code'].isin(countries2drop)) | 
            (df['hs_product_code'].isin(products2drop)))]
```

```python
print(df.shape)
df.ends(2)
```

```python
len(np.unique(df[df['hs_product_code']=="XXXX"].location_code.values))
```

```python
len(np.unique(df.hs_product_code))
```

```python
np.set_printoptions(threshold=1300)
np.unique(df.hs_product_code)
```

```python
# Dropping columns we're not going to use
df = df.dropna(subset=['year', 'export_value', 'export_rca', 'location_code', 'hs_product_code'])

# Droping rows with zero exports, services, and other unspecified products
df = df[((df['export_value']>0) & (~(df['hs_product_code'].isin(['9999', 'XXXX', 'financial', 'ict',
       'transport', 'travel', 'unspecified']))))]
df.shape
```

```python
df.ends(3)
```

# Countries and products should be present in all years

```python
common_country_codes = np.array(list(set(df['location_code'].unique()).intersection(*df.groupby(by=['year'])['location_code'].unique().values))) 
common_product_codes = np.array(list(set(df['hs_product_code'].unique()).intersection(*df.groupby(by=['year'])['hs_product_code'].unique().values))) 
```

```python
(len(common_country_codes), len(common_product_codes))
```

```python
df = df[(df['location_code'].isin(common_country_codes) & df['hs_product_code'].isin(common_product_codes))]
df.shape
```

```python
df.describe(include='all')
```

```python
# checking for missing values
(df.isnull().sum().sum(), df.isna().sum().sum())
```

# Generating new columns

```python
df['tot_exports_by_yr_country'] = df.groupby(['year', 'location_code'])['export_value'].transform('sum')
df['tot_exports_by_yr_product'] = df.groupby(['year', 'hs_product_code'])['export_value'].transform('sum')
```

```python
df.ends(3)
```

```python
(len(df['location_code'].unique()),len(df['hs_product_code'].unique()))
```

```python
df['share'] = np.log(df['export_value'].values)-np.log(df['tot_exports_by_yr_country'].values)
```

```python
df['y'] = -np.log(-df['share'].values)
```

```python
df.ends(3)
```

```python
sns.distplot(df.share.values, rug=True);
```

```python
sns.distplot(df.y.values, rug=True);
```

# Saving data as raw

```python
df.to_csv(PATH_DATA_RAW + 'WTCF_0001_country_hsproduct4digit_{year1}-{year2}.csv'.format(year1=df.year.min(), year2=df.year.max()), 
          index=False)
```
